# **STASTICS BASICS**
## 1. What is statistics, and why is it important?
* Statistics is the science of collecting, analyzing, interpreting, and presenting data.
It is important because it helps in decision-making, identifying trends, making predictions, and understanding patterns in various fields like business, healthcare, sports, government, and research.

##2. What are the two main types of statistics?
* Descriptive statistics: Summarize and describe the features of a dataset.

* Inferential statistics: Make predictions or generalizations about a population based on a sample.

## 3. What are descriptive statistics?
* These are methods for organizing and summarizing data using measures like mean, median, mode, range, variance, and standard deviation.

##4. What is inferential statistics?
* It involves drawing conclusions or making inferences about a population based on data from a sample using techniques like hypothesis testing, confidence intervals, and regression.

##5. What is sampling in statistics?
* Sampling is the process of selecting a subset (sample) from a larger group (population) to analyze and draw conclusions about the entire population.

##6. What are the different types of sampling methods?
* Random Sampling

* Systematic Sampling

* Stratified Sampling

* Cluster Sampling

* Convenience Sampling

* Judgmental/Purposive Sampling

##7. What is the difference between random and non-random sampling?
* Random sampling: Every member has an equal chance of selection (unbiased).

* Non-random sampling: Selection is based on factors like convenience or judgment (may be biased).

##8. Define and give examples of qualitative and quantitative data.
* Qualitative data (categorical): Describes qualities, e.g., gender, color, brand.

* Quantitative data (numerical): Measures quantities, e.g., height, weight, age.

##9. What are the different types of data in statistics?
* Qualitative (categorical)

* Quantitative (numerical)

* Discrete (e.g., number of students)

* Continuous (e.g., temperature)

##10. Explain nominal, ordinal, interval, and ratio levels of measurement.
* Nominal: Categories with no order (e.g., colors)

* Ordinal: Categories with order but no fixed interval (e.g., rankings)

* Interval: Ordered with equal intervals but no true zero (e.g., temperature in °C)

* Ratio: Ordered, equal intervals, and a true zero (e.g., weight, height)

##11. What is the measure of central tendency?
* It indicates the center or typical value of a dataset. The main measures are mean, median, and mode.

##12. Define mean, median, and mode.
* Mean: The average (sum of values ÷ number of values)

* Median: The middle value when data is ordered

* Mode: The most frequently occurring value(s)

##13. What is the significance of the measure of central tendency?
* It helps summarize a large dataset with a single representative value and allows comparison across different datasets.

##15. What is standard deviation, and why is it important?
* Standard deviation is the square root of the variance. It shows how spread out the data is. It’s important for understanding the consistency and volatility of data.

##16. Define and explain the term range in statistics.
* Range is the difference between the highest and lowest values in a dataset.

##17. What is the difference between variance and standard deviation?
* Variance: Average squared deviation from the mean (unit²).

* Standard deviation: Square root of variance (same unit as data).

##18. What is skewness in a dataset?
* Skewness measures the asymmetry of the distribution of values in a dataset.

##19. What does it mean if a dataset is positively or negatively skewed?
* Positive skew: Tail is on the right; mean > median

* Negative skew: Tail is on the left; mean < median

##20. Define and explain kurtosis.
* Kurtosis measures the "tailedness" of a distribution:

* High kurtosis: More outliers, sharper peak

* Low kurtosis: Fewer outliers, flatter peak

##21. What is the purpose of covariance?
* Covariance measures how two variables change together. A positive value indicates that both variables increase together.

##22. What does correlation measure in statistics?
* Correlation quantifies the strength and direction of the relationship between two variables, usually ranging from -1 to +1.

##23. What is the difference between covariance and correlation?
* Covariance: Shows direction but not scale

* Correlation: Standardized measure that shows both direction and strength

##24. What are some real-world applications of statistics?
* Business forecasting

* Medical research

* Government policy-making

* Marketing analysis

* Quality control in manufacturing



````python
#1 How do you calculate the mean, median, and mode of a dataset
import statistics

data = [10, 20, 30, 20, 40, 30, 30]
mean = statistics.mean(data)
median = statistics.median(data)
mode = statistics.mode(data)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
````

````python
#2 Write a Python program to compute the variance and standard deviation of a dataset
variance = statistics.variance(data)
std_dev = statistics.stdev(data)

print("Variance:", variance)
print("Standard Deviation:", std_dev)
````

````python
#4 Implement sampling techniques like random sampling and stratified sampling
import pandas as pd
from sklearn.model_selection import train_test_split
import random

df = pd.DataFrame({
    'Name': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Group': ['X', 'X', 'Y', 'Y', 'X', 'Y', 'X', 'Y']
})
random_sample = df.sample(n=3)
stratified_sample = df.groupby('Group', group_keys=False).apply(lambda x: x.sample(frac=0.5))

print("Random Sample:\n", random_sample)
print("Stratified Sample:\n", stratified_sample)
````

````python
#5 Write a Python function to calculate the range of a dataset
def calculate_range(data):
    return max(data) - min(data)

print("Range:", calculate_range(data))
````

````python
#6 Create a dataset and plot its histogram to visualize skewness
import matplotlib.pyplot as plt
import numpy as np

skewed_data = np.random.exponential(scale=2, size=1000)
plt.hist(skewed_data, bins=30, edgecolor='black')
plt.title("Histogram Showing Skewness")
plt.show()
````

````python
#7 Calculate skewness and kurtosis of a dataset using Python libraries
from scipy.stats import skew, kurtosis

print("Skewness:", skew(skewed_data))
print("Kurtosis:", kurtosis(skewed_data))
````

````python
#12 Implement and compare simple random sampling and systematic sampling
population = list(range(1, 101))
simple_random_sample = random.sample(population, 10)
k = len(population) // 10
systematic_sample = [population[i] for i in range(0, len(population), k)][:10]

print("Simple Random Sample:", simple_random_sample)
print("Systematic Sample:", systematic_sample)
````

````python
#13 Calculate the mean, median, and mode of grouped data
grouped_data = {'class': ['0-10', '10-20', '20-30'], 'frequency': [5, 8, 7]}
df_grouped = pd.DataFrame(grouped_data)
expanded_data = []

for _, row in df_grouped.iterrows():
    lower, upper = map(int, row['class'].split('-'))
    midpoint = (lower + upper) / 2
    expanded_data += [midpoint] * row['frequency']

print("Grouped Data Mean:", statistics.mean(expanded_data))
print("Grouped Data Median:", statistics.median(expanded_data))
print("Grouped Data Mode:", statistics.mode(expanded_data))
````

````python
#14 Simulate data using Python and calculate its central tendency and dispersion
simulated_data = np.random.normal(loc=50, scale=10, size=1000)
print("Simulated Mean:", statistics.mean(simulated_data))
print("Simulated Median:", statistics.median(simulated_data))
print("Simulated Mode:", statistics.mode(simulated_data.round()))
print("Simulated Variance:", statistics.variance(simulated_data))
print("Simulated Standard Deviation:", statistics.stdev(simulated_data))
````

````python
#15 Use NumPy or pandas to summarize a dataset’s descriptive statistics
data = pd.DataFrame({
    'Math': np.random.randint(40, 100, 50),
    'Science': np.random.randint(30, 95, 50),
    'English': np.random.randint(50, 100, 50)
})
print("Descriptive Statistics:\n", data.describe())
````

````python
#16 Plot a boxplot to understand the spread and identify outliers
import seaborn as sns
from scipy.stats import zscore, iqr, skew, kurtosis, pearsonr, spearmanr

plt.figure(figsize=(8, 5))
sns.boxplot(data=data)
plt.title("Boxplot of Subjects")
plt.show()
````

````python
#17 Calculate the interquartile range (IQR) of a dataset
math_scores = data['Math']
math_iqr = iqr(math_scores)
print("Interquartile Range (IQR) for Math:", math_iqr)
````

````python
#18 Implement Z-score normalization and explain its significance
z_scores = zscore(data)
print("Z-score Normalized Data (first 5 rows):\n", pd.DataFrame(z_scores, columns=data.columns).head())
````
